```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'id':[1,2,3,4],'age':[25,30,22,28],'city':['A','B','A','C']}
df = pd.DataFrame(data)
plt.figure()
sns.histplot(df['age'])
plt.title('Distribusi Usia')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='city', y='age', data=df)
plt.title('Boxplot usia per kota')
plt.show()
```
