```python
import pandas as pd
import numpy as np
data = {'id':[1,2,3,4],'age':[25,30,22,28],'city':['A','B','A','C']}
df = pd.DataFrame(data)
# tambahkan baris dengan missing
new = pd.DataFrame({'id':[5],'age':[np.nan],'city':['B']})
df = pd.concat([df,new], ignore_index=True)
print(df)
print('\nMissing per kolom:\n', df.isnull().sum())
```

```python
# imputasi median
df['age'].fillna(df['age'].median(), inplace=True)
# drop duplicates contoh
df = df.drop_duplicates()
# konversi tipe
df['age'] = df['age'].astype(int)
df
```
